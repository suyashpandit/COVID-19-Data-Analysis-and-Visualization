The following measures were created using the DAX query to calculate mean, median, range, standard deviation for new cases and new deaths.
This is universal for other metrics, but just putting the queries I created for new deaths but it applies to all descriptive metrics in the powerBI filed uploaded. 
1. Mean New Deaths = AVERAGE('Covid19GlobalDataCleaned'[New deaths])
2. Median New Deaths = MEDIAN(Covid19GlobalDataCleaned[New deaths])
3. Range New Deaths = MAX('Covid19GlobalDataCleaned'[New deaths]) - MIN('Covid19GlobalDataCleaned'[New deaths])
4. Std Dev New Deaths = STDEV.P('Covid19GlobalDataCleaned'[New deaths])

Similar DAX queries were used to obtain descriptive statistical values for other measures.

I avoided calculating mode because cases and deaths are highly variable and tend to have multiple repeated values but would be of less significance due to the continuous nature of the covid19 pandemic.

Mean would give the average, median per country would be accurate but as I have illustrated in the visualization, using median globally gave me skewed data with median closer to zero because of spikes. 

Global median clearly showed the spikes that was happening during the pandemic. 

Fluctuation was seen through the standard deviation and range showed the difference between the highest cases/deaths and lowest ones.
